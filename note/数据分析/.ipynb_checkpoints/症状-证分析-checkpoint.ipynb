## 中医分析

### 黄疸
> 发病的主要部位在于胆，疾病引发与肝、胆、脾、胃功能失调相关。
> 脾主运化，喜燥恶湿，主升清。
> 胃主受纳腐熟，喜湿恶燥，主降浊。
> 湿邪阻于中焦，使胃的升降失常，脾气不升，则肝气郁结而不能疏泄，
> 胃气不降，可使胆汁疏泄不利，不循常道，两者都能导致胆汁外溢肌肤而发黄。

> 同时，也与体质相关。
> 平素胃热偏盛的人，容易湿从热化，以致阳盛热重，发为阳黄，并有湿和热偏盛之别，
> 又有热重于湿和湿重于热的不同证型。
> 平素脾阳不足，湿可从寒化，以致阴盛寒重，发为阴黄。
