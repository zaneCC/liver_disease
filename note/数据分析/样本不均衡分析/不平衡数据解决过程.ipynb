### 随机重采样建立多个分类器

两类样本：1:阳黄证，2:阳阴黄证

使用分类器：100个决策树

采样方法：（按以下顺序执行）
1. 打乱数据
2. 分别从1类和2类中抽取15条数据（共30条）
3. 对这30条数据建立一棵决策树（共建立100棵）
4. 对100棵决策树进行训练，得到训练模型
5. 对100棵决策树进行预测，得到预测结果集
6. 对预测结果集中取最多数，作为最终结果

### SMOTE & SMOTE Borderline 合成采样分析

1. 从整体来看，SMOTE Borderline性能优于 SMOTE。
2. 两者都相应提升了第2类的分类性能，同时可见第2类分类效果基本都优于第1类，是因为算法本身更强调第2类分类效果（可能是通常应用中需要更强调异常类情况）。（其原理呢？）
3. 两者都是采用 knn 随机选取少数类样本进行新样本合成，因此产生的新样本具有随机性，导致样本合成结果时好时坏。
